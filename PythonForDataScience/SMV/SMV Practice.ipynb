## Support Machine Vectors

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df_feat = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df_feat.head(2)
```

```python
#Skip visualizations for this one
```

```python
from sklearn.model_selection import train_test_split
```

```python
X =  df_feat
y = cancer['target']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
#Initialize the model
model = SVC()
```

```python
#fIT THE MODEL WITH TRAINING DATA
model.fit(X_train,y_train)
```

```python
#Allow our model to make predictions based on our testing data
predictions = model.predict(X_test)
```

```python
#Print the reports to get more info about our model including the accuracy - This compares how far the predictions are from our Y_test
from sklearn.metrics import classification_report, confusion_matrix


print(confusion_matrix(y_test, predictions))
print('\n')
print(classification_report(y_test, predictions))
```

```python
#In the video, he got the wrong reportwith ill-fefined Precision and F-Score
#Following along- we'll use a method called grid search to find the right parameters
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#We use the actual parameters that go into the SVC Model
# C - controls the cost of misclassification on the training data - Large C value gives us low vias and highvarience as we penalize the cost
param_grid = {'C':[0.1,1,10,100,1000], 'gamma':[1,0.1,0.01,0.001,0.0001]}
```

```python
grid = GridSearchCV(SVC(), param_grid, verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test, grid_predictions))
print('\n')
print(classification_report(y_test, grid_predictions))
```
